```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rc('xtick', labelsize=16) 
matplotlib.rc('ytick', labelsize=16) 
%matplotlib inline
```

```python
## RoBERTa
con_indus = [77.81, 78.81, 78.45, 80.55]
con_tec = [75.54, 79.23, 77.46, 77.93]
indus_con= [72.49, 73.82, 77.09, 75.07]
indus_tec = [73.99, 77.50, 78.50, 76.70]
tec_con = [74.63, 75.72, 77.93, 77.46]
tec_indus = [77.91, 76.71, 80.05, 79.00]
```

```python
x = ['RoBERTa', 'RoBERTa\n+LPFT', 'RoBERTa\n+Prompt', 'RoBERTa+\nPrompt+LPFT']
plt.figure(figsize=(9,5))
plt.plot(x, con_indus, label='Con-->Ind', marker='o')
plt.plot(x, con_tec, label='Con-->Tec', marker='o')
plt.plot(x, indus_con, label='Ind-->Con', marker='o')
plt.plot(x, indus_tec, label='Ind-->Tec', marker='o')
plt.plot(x, tec_con, label='Tec-->Con', marker='o')
plt.plot(x, tec_indus, label='Tec-->Ind', marker='o')
plt.ylabel('F1', fontsize=16)
plt.xlabel('Model', fontsize=16)
plt.legend(loc=4)
plt.show()
```
